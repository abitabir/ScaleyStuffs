```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("kmeans.csv")
df
```

```python
coordinates = df[["weight", "height"]].values.tolist()
coordinates
X = []
Y = []
for element in coordinates:
    X.append(element[0])
    Y.append(element[1])
```

```python
number_of_clusters = 3
k = KMeans(n_clusters=number_of_clusters)
clustered_data = k.fit_predict(coordinates)
clustered_data
```

```python
cluster_one = []
cluster_two = []
cluster_three = []
for i in range(len(clustered_data)):
    if clustered_data[i] == 0:
        cluster_one.append(coordinates[i])
    elif clustered_data[i] == 1:
        cluster_two.append(coordinates[i])
    else:
        cluster_three.append(coordinates[i])
```

```python
print("As per Question 5 (i) ~ ")
print("cluster one:\n", cluster_one)
print("\n")
print("cluster two:\n", cluster_two)
print("\n")
print("cluster three:\n", cluster_three)
```

```python

import matplotlib.pyplot as plt
plt.scatter(X, Y, c=clustered_data, s=len(coordinates), cmap="viridis")
```

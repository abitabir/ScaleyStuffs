```python
import pandas as pd
import numpy as np
import scipy.stats as stats
from scipy.stats import shapiro

df = pd.read_csv("survey_responses.csv")
df
```

```python
quality_of_sleep_by_gender = df[["gender", "qualslp"]].values.tolist()
quality_of_sleep_by_gender
gender = df["gender"].values.tolist()
quality_sleep = df["qualslp"].values.tolist()
```

```python
shapiro(gender)
shapiro(quality_sleep)
# since both p values are > 0.05, these do follow a normal distribution thus we will use
# the parametric test – t test – to get correlation value.
stats.spearmanr(gender, quality_sleep)
# if p value < 0.05, there is a significant correlation between gender and quality of sleep
```

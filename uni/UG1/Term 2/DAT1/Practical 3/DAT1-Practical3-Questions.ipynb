** T11) Is there any missing data? Is this a messy or a clean dataset?**

```python
# Write your answer here
# we either replace missing values, or we delete all the records containing missing values for clean stats
# df.dropna()  # retruns database with all records with missing values in original data set dropped; can take axis parameter
# if axis="columns"/1 you drop the columns, if it is set to "index"/0 (as default) i.e. rows,
# the records containing the missing datas are lost
df.isnull()  # returns database with True or False booleans having checked for missing data values
# (True if there is no value provided for a specific records's column data)
df.isnull().sum()  # getting all missing values summed for every column (going across axis=0 default)
# note: .sum() method not .sum attribute
columns_number = df.sum().count()  # calculates and assigns number of column_headers(/attributes ~ whose values could in turn be inputted for each record/instance)

# in pandas docs is explicitly written isnull = isna  # they do the same thing so are interchangable
missing_data_summary = duplicate_free_df.isna().sum()
missing_data_summary_nice_layout = missing_data_summary.sort_values(ascending=False)
# gets the missing data and sorts in descending order
missing_data_summary_nice_layout  # returns the summary that is nice

nonnull_counts_in_columns_series = duplicate_free_df.isnull().count()  # returns series of the column headers and the counts of
# non-NA or nonnull values in them (adding up the number of records vertically for every column column - as axis parameter
# defaults to 0/index {basically think y = 0 is same as x axis mindbogglingnesses})
# basically has 10358 for every record in series (ie db column ~ conversions) so this works too in a bad practice way tho
# missing_data_percentage_series = missing_data_summary/10358
missing_data_percentage_series = missing_data_summary/nonnull_counts_in_columns_series   # dividing the values of the Series
# the number of records with missing data (for every column) by the number of records in total (for that column)
missing_data_percentages_nice_layout = (missing_data_percentage_series).sort_values(ascending=False)

# shortwinded route with less variables and less understanding but less confusion maybe:
# missingDataPercentage = (duplicate_free_df.isnull().sum()/duplicate_free_df.isnull().count()).sort_values(ascending=False)

missing_data_percentages_nice_layout

# concatenating the two Series
missing_data = pd.concat([missing_data_summary_nice_layout, missing_data_percentages_nice_layout], axis=1, keys=['Missing', 'Percentage'])
missing_data  # nice db returned


```

** T12) Impute the categorical columns with the mode value**

```python
# note: categorical variables (columns) include data which may be divided into further groups (e.g. gender)
clean_df = duplicate_free_df.copy()  # maintaining differnt iterations of dataframe so we can backtrack
# replacing missing data values in the categorical columns: "Type", "Content Rating", "Android Version" and
# "Current Version" with the mode of the respective column that the missing data value is in
clean_df["TYPE"].fillna(value=clean_df["Type".upper()].mode()[0], inplace=True)  # note: my df's columnnames are in caps
clean_df["CONTENT RATING"].fillna(value=clean_df["CONTENT RATING"].mode()[0], inplace=True)
clean_df["ANDROID VERSION"].fillna(value=clean_df["ANDROID VERSION"].mode()[0], inplace=True)
clean_df["CURRENT VERSION"].fillna(value=clean_df["CURRENT VERSION"].mode()[0], inplace=True)


# well, 'clean', but could be inaccurate tbf
```
